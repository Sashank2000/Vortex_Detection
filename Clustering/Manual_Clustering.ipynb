### Magnitude Based Clustering

```python
import math
mag = np.empty((128,128))

m,n,o = 0,0,0
for i in range(128):
    for j in range(128):
        p = c1[i][j]
        q = c2[i][j]
        r = c3[i][j]
        mag[i][j] = math.sqrt((p*p+q*q+r*r))
```

```python
mag_1d = mag.ravel()
mag_1d

mag_max = mag_1d.max()
print(mag_max)
mag_min = mag_1d.min()
print(mag_min)
```

```python
mag_1 = []
mag_2 = []
mag_3 = []
mag_4 = []
mag_5 = []

diff = mag_max-mag_min
diff = diff/5

for i in range(128):
    for j in range(128):
        if mag[i][j]<mag_min+diff:
            mag_1.append([i,j])
        elif mag[i][j]<mag_min+(2*diff) and mag[i][j]>= mag_min+diff:
            mag_2.append([i,j])
        elif mag[i][j]< mag_min+(3*diff) and mag[i][j]>= mag_min+(2*diff):
            mag_3.append([i,j])
        elif mag[i][j]< mag_min+(4*diff) and mag[i][j]>= mag_min+(3*diff):
            mag_4.append([i,j])
        else:
            mag_5.append([i,j])
```

```python
points=[]
for i in range(0,2):
    temp=[]
    for j in mag_1:
        temp.append(j[i])
    points.append(temp) 
    
for i in range(0,2):
    temp=[]
    for j in mag_2:
        temp.append(j[i])
    points.append(temp)
    
for i in range(0,2):
    temp=[]
    for j in mag_3:
        temp.append(j[i])
    points.append(temp)
    
for i in range(0,2):
    temp=[]
    for j in mag_4:
        temp.append(j[i])
    points.append(temp)

for i in range(0,2):
    temp=[]
    for j in mag_5:
        temp.append(j[i])
    points.append(temp)
    
print(len(points))
```

```python
import matplotlib.pyplot as plt

label = [0,1,2,3,4]
for i in range(0,10,2):
    plt.scatter(points[i],points[i+1],label = label[i//2])
plt.legend()
plt.savefig('/home/sashankk/Proj/MAGNETOHYDRODYNAMIC-TURBULENCE-master/sliced/clusters/100007_mag.png')
plt.show()
```

### Direction based Clustering

```python
import math
mag = np.empty((128,128))

for i in range(128):
    for j in range(128):
        p = c1[i][j]
        q = c2[i][j]
        r = c3[i][j]
        mod = math.sqrt((q*q+r*r))
        temp1 = math.acos(q/mod)
#         mag[i][j] = math.degrees(temp1)
        temp2 = math.acos(r/mod)
        temp2 = math.degrees(temp2)
        if temp2>90:
            mag[i][j] = 360 - math.degrees(temp1)
        else:
            mag[i][j] = math.degrees(temp1)
```

```python
mag_1d = mag.ravel()
mag_1d

mag_max = mag_1d.max()
print(mag_max)
mag_min = mag_1d.min()
print(mag_min)
```

```python
mag_1 = []
mag_2 = []
mag_3 = []
mag_4 = []
mag_5 = []
mag_6 = []
mag_7 = []
mag_8 = []

diff = mag_max-mag_min
diff = diff/6
print(diff)

for i in range(128):
    for j in range(128):
        if mag[i][j]<mag_min+diff:
            mag_1.append([i,j])
        elif mag[i][j]<mag_min+(2*diff) and mag[i][j]>= mag_min+diff:
            mag_2.append([i,j])
        elif mag[i][j]< mag_min+(3*diff) and mag[i][j]>= mag_min+(2*diff):
            mag_3.append([i,j])
        elif mag[i][j]< mag_min+(4*diff) and mag[i][j]>= mag_min+(3*diff):
            mag_4.append([i,j])
        elif mag[i][j]< mag_min+(5*diff) and mag[i][j]>= mag_min+(4*diff):
            mag_5.append([i,j])
        elif mag[i][j]< mag_min+(6*diff) and mag[i][j]>= mag_min+(5*diff):
            mag_6.append([i,j])
        elif mag[i][j]< mag_min+(7*diff) and mag[i][j]>= mag_min+(6*diff):
            mag_7.append([i,j])
        else:
            mag_8.append([i,j])
```

```python
points=[]
for i in range(0,2):
    temp=[]
    for j in mag_1:
        temp.append(j[i])
    points.append(temp) 
    
for i in range(0,2):
    temp=[]
    for j in mag_2:
        temp.append(j[i])
    points.append(temp)
    
for i in range(0,2):
    temp=[]
    for j in mag_3:
        temp.append(j[i])
    points.append(temp)
    
for i in range(0,2):
    temp=[]
    for j in mag_4:
        temp.append(j[i])
    points.append(temp)

for i in range(0,2):
    temp=[]
    for j in mag_5:
        temp.append(j[i])
    points.append(temp)

for i in range(0,2):
    temp=[]
    for j in mag_6:
        temp.append(j[i])
    points.append(temp)
    
for i in range(0,2):
    temp=[]
    for j in mag_7:
        temp.append(j[i])
    points.append(temp)
    
for i in range(0,2):
    temp=[]
    for j in mag_8:
        temp.append(j[i])
    points.append(temp)
    
print(len(points))
```
