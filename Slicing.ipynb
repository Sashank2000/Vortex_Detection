# Slicing One Timeframe

```python
import h5py
import numpy as np
import pandas as pd
```

```python
f = h5py.File("/home/sashankk/Proj/maxhelmaxrot/data/velocity.100000.h5","r")
```

```python
list(f.keys())
```

```python
ds=f["PS"]
list(ds.keys())
```

```python
c1=ds["vx"]
c2=ds["vy"]
c3=ds["vz"]
print(c2[0][0])
print(c3[0][0])
```

```python
c1[0][0]
```

```python
itr=0            #Keeping x - co ordinate constant to visualize each 2-D slice of the 3-D image.
while itr<128:
    write_vx=np.empty((128,128))

    i,j,k=itr,0,0
    for j in range(len(c1[i])):
        for k in range(len(c1[i][j])):
            write_vx[j][k]=c1[i][j][k]

    write_vy=np.empty((128,128))
    i,j,k=itr,0,0
    for j in range(len(c2[i])):
        for k in range(len(c2[i][j])):
            write_vy[j][k]=c2[i][j][k]


    write_vz=np.empty((128,128))
    i,j,k=itr,0,0
    for j in range(len(c3[i])):
        for k in range(len(c3[i][j])):
            write_vz[j][k]=c3[i][j][k]
            
    filename="velocity10000"+str(itr)+".h5"
    filepath="/home/sashankk/Proj/maxhelmaxrot/data_vz2/"+filename
    with h5py.File(filepath,"w") as hf:
        g1=hf.create_group("PS")
        g1.create_dataset("vx", data=write_vx)
        g1.create_dataset("vy", data=write_vy)
        g1.create_dataset("vz", data=write_vz)
    
    itr+=1
    
```
